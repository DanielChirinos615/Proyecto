```python
%%writefile r1.cpp
#include <iostream>
int termino (int t){
    if (t <= 2)
        return 1;
    else
    return termino(t-1)+termino(t-2);
}
void fibonacci(int n){
    if(n>0){
        fibonacci(n-1);
        std::cout<<termino(n)<<" ";
    }
}
int main (){
    fibonacci(5);
}
```

```python
%%writefile n1.cpp
#include<iostream>
using namespace std;
void columnas(int n){
  if(n>0){
    columnas(n-1);
    cout<<"*";
  }
  else  
  cout<<endl;
}
void filas(int n){
    if(n==0)return ;
    else{
      filas(n-1);
      columnas(n);}
}
int main(){
  filas(5);
}
```

```python
!g++ n1.cpp && ./a.out
```
